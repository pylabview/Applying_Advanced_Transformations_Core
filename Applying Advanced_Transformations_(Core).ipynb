### Splitting Hero|Publisher column

```python
## Testing splitting
df_info['Hero|Publisher'].str.split('|',expand=True)
```

```python
## save the 2 new columns into the dataframe
df_info[['hero_names','publisher']] = df_info['Hero|Publisher'].str.split('|', expand=True)
df_info.head(2)
```

```python
#Dropping column
df_info.drop(['Hero|Publisher'], axis=1, inplace=True)
```

```python
#Check again info
df_info.info()
```

### Converting Measurements

```python
#Checking Measuremnts string
df_info.loc[0,'Measurements']
```

```python
# Doing some transformations: replace double quoutetions and removing units from weight and height
mm = df_info.loc[0,'Measurements'].replace("'", '"').replace(" cm","").replace(" kg","")
mm
```

```python
# Checking dict convertion
json.loads(mm)
```

>Applying this to the entire column

```python
df_info['Measurements'] = df_info['Measurements'].str.replace("'", '"').str.replace(" cm","").str.replace(" kg","")
df_info['Measurements'] = df_info['Measurements'].apply(json.loads)
df_info.head()
```

```python
h_w = df_info['Measurements'].apply(pd.Series)
```

```python
h_w
```

### Concatening height and weight

```python
df_info = pd.concat([df_info,h_w],axis=1)
df_info.head()
```

```python
df_info.drop(columns=['Measurements'],inplace=True)
```

```python
#Converting Height to float
df_info['Height'] = df_info['Height'].astype(float)
```

```python
#Converting Weight to float
df_info['Weight'] = df_info['Weight'].astype(float)
```

```python
df_info.info()
```

```python
#Exploring one Powers single value
df_powers.loc[10,'Powers'].strip().replace("  ","").split(",")
```

```python
#Convert string to list
df_powers['Powers'] = df_powers['Powers'].str.strip().replace("  ","")
```

```python
df_powers.head()
```

```python
df_powers['Powers'] = df_powers['Powers'].str.split(",")
```

#### Get a list of all options using df.explode

```python
df_powers['Powers'].value_counts()
```

```python
## exploding the column of lists
exploded = df_powers.explode('Powers')
```

```python
## saving the unique values from the exploded column
cols_to_make = exploded['Powers'].dropna().unique()
cols_to_make
```

```python
df_powers
```

```python
#one-hot-encoded columns for every power that appears in the dataset
for i,col in enumerate(cols_to_make):
    df_powers[col] = col in df_powers.loc[i,'Powers']
df_powers.head()

    
df_powers.head()
```

```python
# Droping Powers column
df_powers.drop(columns=['Powers'],inplace=True)
```
